```python
import pandas as pd
import json
import folium
```

```python
df = pd.read_csv("scanned_aps_sanitized.csv", usecols=["created_at", "ssid", "bssid", "level", "latitude", "longitude"])
df["created_at"] = pd.to_datetime(df["created_at"], format="ISO8601")

df = df.loc[(df["latitude"] < 47.552425) & (df["latitude"] >= 47.550714) & (df["longitude"] >= -122.266685 ) & (df["longitude"] < -122.262940)]
# df = df.loc[(df["created_at"] >= "2024-05-01") & (df["created_at"] < "2024-06-01")]


## Michael's place -- box (47.552425, -122.266685) (47.550714, -122.262940)
df
```

```python
df.loc[df["ssid"] == "Bert"]["bssid"].unique()
```

```python
m = folium.Map(location=[40, -99], zoom_start=4)
# More than 10k will freeze leaflet.
grouped = df.loc[df["ssid"].isin(["Ernie", "Bert"])].groupby(["latitude", "longitude", "bssid", "ssid"]).agg({"level": lambda x: ', '.join(x.astype(str))})

print(grouped.shape)

for (latitude, longitude, bssid, ssid), row in grouped.iterrows():
    folium.Marker(
        location=[latitude, longitude],
        tooltip=f"{ssid} - {bssid}",
        popup=f'level: {row["level"]}',
        icon=folium.Icon(color="green")
    ).add_to(m)

m
```
